```python
import numpy as np
import pandas as pd
from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv('./course-files/marathon_results_2016.csv', usecols = ['Bib','40K','Half','Pace','Age','M/F','Country','State','City'], index_col = 'Bib')
```

```python
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
```

```python
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
```

```python
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df.head()
```

```python
df.pivot_table(index='Age', columns = 'M/F', values = 'TotalSeconds', aggfunc = 'mean')
```

```python
df.pivot_table(index='Age', columns = 'M/F', values = ['TotalSeconds','HalfSeconds'], aggfunc = 'mean')
```

```python
df.pivot_table(index='Age', columns = 'M/F', values = ['TotalSeconds','HalfSeconds'], aggfunc = 'mean').swaplevel(axis=1).head()
```

```python
df.pivot_table(index='Age', columns = 'M/F', values = ['TotalSeconds','HalfSeconds'], aggfunc = 'mean').swaplevel(axis=1).sort_index(axis=1).head()
```

```python
df.pivot_table(index='Age', columns = 'M/F', values = 'State', aggfunc = 'count')
```

```python
df.pivot_table(index='Age', columns = 'Country', values = 'State', aggfunc = 'count')
```

```python
df.pivot_table(index='Country',values = 'State',aggfunc = 'count')
```

```python
df.pivot_table(index='Country',values = 'City',aggfunc = 'count').sort_values('City', ascending = False)
```

```python
df.pivot_table(index = ['Country','State','City'], values = 'TotalSeconds', aggfunc = 'mean' ).sort_values('TotalSeconds', ascending = False)
```
